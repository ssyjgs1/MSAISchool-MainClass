```python
# 필요한 기본 패키지 추가
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
data = iris.data
label = iris.target
columns = iris.feature_names
```

```python
data = pd.DataFrame(data, columns=columns) # 데이터프레임화
data.head() # 앞 5개 정보 열람
```

```python
data.shape
```

```python
# 데이터의 준비
X_train, X_test, y_train, y_test = train_test_split(data,label,test_size=0.2,random_state=2022)
# Logistic Regression (Classification의 일종. 이름에 낚이지 말자)
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score # 정확도 판단 지표
```

```python
print('로지스틱 회귀, 정확도 : {:.2f}'.format(accuracy_score(y_test,y_pred)))
```

```python
# Support Vector Machine
from sklearn.svm import SVC
svc = SVC(C=1000)
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
print('Support Vector Machine 정확도 : {:.2f}'.format(accuracy_score(y_test,y_pred)))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=100)
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
print('Decision Tree 정확도 : {:.2f}'.format(accuracy_score(y_test,y_pred)))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=5)
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
print('Random Forest 정확도 : {:.2f}'.format(accuracy_score(y_test,y_pred)))
```
