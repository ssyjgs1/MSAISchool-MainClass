```python
# 필요한 패키지 불러오기
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') # 스타일 지정
```

```python
# 1차함수 만들기
def linear_function(x):
    a = 0.8 # 기울기
    b = 2
    return a * x + b
```

```python
print(linear_function(5))
```

```python
# x값을 배열로 만들고 던지고 y값을 받아내서 그래프를 그려보자(시각화)
# -5 ~ 5까지 0.1 단위로 만든다
x = np.arange(-5,5,0.1)
y = linear_function(x) # 배열을 주면 그 배열의 요소 갯수만큼 함수를 호출하고 배열로 반환해 줌
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
# 2차함수
def quadratic_function(x):
    a = 1
    b = -1
    c = -2

    return a*x**2 + b*x + c
```

```python
print(quadratic_function(2))
```

```python
y = quadratic_function(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

```python
# 3차함수(다항함수)
def cubic_function(x):
    a=4
    b=0
    c=-1
    d=-8
    return a*x**3 + b*x**2 + c*x + d
```

```python
y = cubic_function(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
```

```python
# 최소값 구하기
def my_func(x):
    a=1
    b=-3
    c=10
    return a*x**2 + b*x + c # 2차함수식
```

```python
x = np.arange(-10,10,0.1)
y = my_func(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('my func')
plt.scatter(1.5,my_func(1.5))
```

```python
# 지수함수
def exponential_function(x):
    a=4
    return a**x
```

```python
print(exponential_function(4))
print(exponential_function(0))
```

```python
x = np.arange(-3,2,0.1)
y = exponential_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('exponential_function')
plt.xlim(-4,3)
plt.ylim(-1,15)
```
