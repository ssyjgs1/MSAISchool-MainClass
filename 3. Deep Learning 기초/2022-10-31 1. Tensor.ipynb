```python
# 필요한 패키지 불러오기
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') # 스타일 지정
```

```python
# 0차원 tensor. scala 값(=방향성 없음)
x = np.array(3)
print(x)
print(x.shape) # 0차원이라 모양이 없이 비어있는 형태를 띰
print(np.ndim(x)) # 몇 차원인지
```

```python
# 1차원 tensor. 방향이 있는 벡터
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
c = a + b # 같은 위치에 있는 애들끼리 연산됨

print(c)
print(c.shape)
print(np.ndim(c))
```

```python
c = a * b
print(c)
print(c.shape)
print(np.ndim(c))
```

```python
# scala와 vector의 곱
a = np.array(10) # scala
b = np.array([1,2,3]) # 1차원 tensor
c = a * b
print(c)
print(c.shape)
print(np.ndim(c))
```

```python
# 전치행렬
A = np.array([[1,2,3],[4,5,6]]) # 2차원 tensor
print('A\n', A)
print('A.shape\n', A.shape)
print('A.dim\n', np.ndim(A))
print('--------------------')
```

```python
A_ = A.T # T에 전치행렬 기능이 들어가있음
print('A_\n', A_)
print('A_.shape\n', A_.shape)
print('A_.dim\n', np.ndim(A_))
print('--------------------')
```

```python
# 번외편 - 지난 수업 과제: PKL(모델의 내보내는 형식 중 하나) 파일 저장 및 로딩
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
import joblib

dataset = datasets.load_iris()

X, y = dataset['data'], dataset['target']
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

joblib.dump(model,'./knn_model.pkl') # 모델 저장해서 현재 경로로 내보내기

loaded_model = joblib.load('./knn_model.pkl') # 모델 불러오기
# loaded_model.predict

score = loaded_model.score(X,y)
print('정확도 : {score:.3f}'.format(score=score))
```
