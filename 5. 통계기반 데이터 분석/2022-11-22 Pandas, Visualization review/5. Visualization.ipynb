# 바 그래프

```python
data = {
    '바나나':3,'사과':2,'오렌지':10
}

# 선 그래프 : plot, 세로바 그래프 : bar, 가로바 그래프 : barh
plt.bar([0,1,2],data.values())
plt.title('과일 개수')
plt.xticks([0,1,2],data.keys())
plt.yticks(sorted(data.values()))
plt.grid(True)
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
```

# 파이차트

```python
ratio = [34,32,16,18]
labels=['apple','banana','melon','grape']
explode=[0.05, 0.05,0.05,0.05]
colors=['silver','gold','whitesmoke','lightgray']

# 파이차트 그리는 메소드 plt.pie
plt.pie(ratio,labels=labels,autopct='%.1f%%', #autopct 소수점 자리 표기 관련
        startangle=260,counterclock=False, shadow=True, explode=explode,
        colors=colors) # explode는 그래프를 구성하는 칸이 떨어져 있는지 여부
plt.show()
```

```python
wedgeprops = {'width':0.7,'linewidth':5}
plt.pie(ratio,labels=labels,autopct='%.1f%%', #autopct 소수점 자리 표기 관련
        startangle=260,counterclock=False, shadow=True,
        colors=colors,wedgeprops=wedgeprops)
```

# 박스 그래프

```python
np.random.seed(0)

# 1. 기본 스타일 설정
plt.style.use('default')
# plt.rcParams['figure.figsize'] = (4, 3)
# plt.rcParams['font.size'] = 12

# 2. 데이터 준비
'''
np.random.normal -> 정규분포 데이터 생성
mu : 중앙값
sigma : 표준편차
size : 결과 사이즈
'''
# 중앙값이 0이고 표준편차가 2인 랜덤 데이터 1000개 생성
data_a = np.random.normal(0, 2.0, 1000)
# 중앙값이 -3이고 표준편차가 1.5인 랜덤 데이터 500개 생성
data_b = np.random.normal(-3.0, 1.5, 500)
# 중앙값이 1.2이고 표준편차가 1.5인 랜덤 데이터 1500개 생성
data_c = np.random.normal(1.2, 1.5, 1500)
# 3. 그래프 그리기
fig, ax = plt.subplots()
'''
notch : bool (중앙값의 95%신뢰구간을 노치형태로 나타냄)
whis : float (수염의 길이를 정할 때 IQR에 곱해지는 변수 default : 1.5)
'''

# 박스 그래프 : boxplot
# notch : 신뢰구간(95%)
ax.boxplot([data_a,data_b,data_c],notch=True,whis=1.5,vert=True)
ax.set_ylim(-10.0,10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

```python
# 가로 박스 그래프 그리기
fig, ax = plt.subplots()
ax.boxplot([data_a,data_b,data_c],notch=True,whis=1.5,vert=False)
ax.set_xlim(-10.0,10.0)
plt.show()
```

# 바이올린 플롯(violin plot)

```python
plt.style.use('default')

# 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 바이올린플롯 : plt.violinplot
violin=plt.violinplot([data_a,data_b,data_c],showmeans=True, showextrema=True,
                     showmedians=True)
# showmeans : 평균값 줄쳐서 보여주기
# showextrema : 최댓값, 최솟값 위치 줄쳐서 보여주기
# showmedians : 중앙값 위치 줄쳐서 보여주기
plt.xticks(np.arange(1,4),labels=['A','B','C'])
plt.show()
```

```python
plt.style.use('default')

# 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 바이올린플롯 : plt.violinplot
violin=plt.violinplot([data_a,data_b,data_c],showmeans=True, showextrema=True,
                     showmedians=True)
# showmeans : 평균값 줄쳐서 보여주기
# showextrema : 최댓값, 최솟값 위치 줄쳐서 보여주기
# showmedians : 중앙값 위치 줄쳐서 보여주기
plt.xticks(np.arange(1,4),labels=['A','B','C'])
violin['bodies'][0].set_facecolor('red')
violin['cbars'].set_edgecolor('red')
violin['cmaxes'].set_edgecolor('orange')
violin['cmins'].set_edgecolor('black')
violin['cmeans'].set_edgecolor('red')
violin['cmedians'].set_edgecolor('black')

plt.show()
```
