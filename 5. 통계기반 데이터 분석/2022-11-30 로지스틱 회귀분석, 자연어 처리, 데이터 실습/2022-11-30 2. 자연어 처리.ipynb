```python
import nltk
nltk.download('book',quiet=True)
```

```python
nltk.corpus.gutenberg.fileids()
```

```python
emma_raw = nltk.corpus.gutenberg.raw('austen-emma.txt')
```

```python
print(emma_raw[:1000])
```

### 토큰 생성

```python
from nltk.tokenize import sent_tokenize # 문장 기준으로 토큰화함
print(sent_tokenize(emma_raw[:1000])[0])
print(sent_tokenize(emma_raw[:1000])[1])
```

```python
from nltk.tokenize import word_tokenize # 단어 기준으로 토큰화함함
word_tokenize(emma_raw[:1000])
```

```python
from nltk.tokenize import RegexpTokenizer # 정규표현식으로 토큰화함(쉼표나 대괄호 같은 거 빠짐)
retokenize = RegexpTokenizer('[\w]+')
retokenize.tokenize(emma_raw[:1000])
```

### 형태소 분석

```python
from nltk.stem import PorterStemmer, LancasterStemmer

st1 = PorterStemmer()
st2 = LancasterStemmer()

words = ['fly','flies','flying','flew','flown']

print([st1.stem(w) for w in words])
print([st2.stem(w) for w in words])
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4')

lm = WordNetLemmatizer()

[lm.lemmatize(w, pos='v') for w in words]
```

### 품사 부착
- NNP : 단수 고유명사
- VB : 동사
- VBP : 동사 현재형

```python
nltk.help.upenn_tagset('VB')
```

```python
from nltk.tag import pos_tag
tagged_list = pos_tag(word_tokenize(emma_raw[:1000]))
```

```python
[i[0] for i in tagged_list if i[1]=='NN']
```

```python
from nltk.tag import untag
untag(tagged_list)
```

```python
from nltk import Text
text = Text(retokenize.tokenize(emma_raw))
```

```python
import matplotlib.pyplot as plt
text.plot(30) # 자주 나오는 상위 30개 단어 보여줌
plt.show()
```

```python
text.dispersion_plot(['Emma','Knightley','Frank','so']) # 단어가 나오는 위치 확인
```

```python
text.concordance('Emma')
```

```python
text.similar('Emma')
```

```python
# 빈도수 확인
fd = text.vocab()
type(fd)
```

```python
from nltk import FreqDist

stopwords=['Mr','Mrs','Miss','Mr','Ms','Deer']
emma_tokens = pos_tag(retokenize.tokenize(emma_raw))
names_list = [t[0] for t in emma_tokens if t[1] == 'NNP' and t[0] not in stopwords]
fd_names =FreqDist(names_list)
```

```python
fd_names.N(), fd_names['Emma'], fd_names.freq('Emma')
# 전체 이름 수, Emma 등장 수, Emma 등장 확률
```

```python
names_list
```

```python
fd_names.most_common(5) # 가장 많이 등장하는 5개 이름
```

```python
from wordcloud import WordCloud
wc = WordCloud(width=2000, height=1200, background_color='white')
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.show()
```
