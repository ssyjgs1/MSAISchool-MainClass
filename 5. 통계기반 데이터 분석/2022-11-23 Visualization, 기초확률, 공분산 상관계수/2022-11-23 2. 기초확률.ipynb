# **확률**
- 사건이 일어날 확률
- 확률의 고전적인 정의
    - 실험 결과가 N가지 일 때, 일어난 사건 A가 m번 일어났으면 P(A) = m/N
    - 단, 각 실험 가능성은 모두 같을 때
    - ex) 주사위{1, 2, 3, 4, 5, 6} --> 1/6 확률
- 확률의 경험적인 정의
    - 반복 횟수를 N이라 했을 때, 무한번 반복했을 때 발생하는 A의 횟수를 m이라 할 때, P(A) = lim m/N
- 확률의 공리적인 정의
    - 콜모코로프 공리적 정의 3가지
        1. 0 ≤ P(A) ≤ 1
        2. 표본공간 P(S) = 1
        3. 실험이 표본공간의 부분집합이면서 다른 사건과 교집합이 없을 때각 실험의 확률을 더한 값과 합집합의 확률은 같다.
        ex) 동전 던지기를 할 때, 앞면 : H, 뒷면 : T, 표본공간 = S{HH, HT, TT, TH}, 앞이 안 나오는 사건(A{TT}) 과 뒷면이 안 나올 사건(B{HH})면 A와 B는 교집합이 없습니다.
        
    P(A∪B) = P({TT,HH}) = P({TT}) + P({HH})

    P(A∪B) = P(A) + P(B)


# 확률의 성질
- 공집합의 확률 = 0
- 여집합의 확률
      실험 A의 여집합은 1 - P(A)
- 포함-배제 원리
      P(A∪B) = P(A) + P(B) - P(A∩B)
- 각 사건이 배반사상(event)일 경우
      P(A1∪A2∪A3.....∪An) = P(A1)+P(A2)....P(An)
      A가 B의 부분집합일 때 P(A) ≤ P(B)


## 조건부 확률
- 어떤 사건 B가 일어날 때(P(B) > 0) 사건 A가 일어날 확률

$$P(A|B) = \frac{P(A\cap B)}{P(B)}$$

- 사건 A가 발생했을 때 사건 B가 발생할 확률
$$P(B|A) = \frac{P(A\cap B)}{P(A)}$$

$$P(A\cap B) = P(A|B)P(B) = P(B|A)P(A)$$

- 독립사건
$$P(A|B) = P(A)$$
$$P(B|A) = B(B)$$

## 베이즈 정리

- 분할사건
      사건 K개가 일어났을 때 A1 ... Ak 서로 다 교집합이 없는경우 분할 사건
      사건 K개를 모두 합집합했을 때 S가 되는 경우

$$ B = (B \cap {A_1}) U ⋅⋅⋅ U (B \cap {A_k}) $$



$$ P(B) = P(B \cap {A_1}) + ⋅ ⋅ ⋅ + P(B ∩ {A_k}) $$



$$P({A_i}|B) = \frac{P(B|{A_i})P({A_i})}{P(B)} = \frac {P(B|{A_i})P({B_i})}{P(B|{A_1})P({A_1}) + P(B|{A_2})P({A_2})+ ⋅ ⋅ ⋅+ P(B|{A_k})P({A_k})}$$

- 사전확률을 사후확률로 추정하는 정리

## 확률 변수와 확률 분포

- 확률변수
      표본공간에서 정의된 실수값 함수
      확률 변수들의 통계량 찾고 
      확률 변수를 모수를 추정
      대문자 X로 표현

- 확률 분포
      확률 변수 값과 확률을 대응시키는 표, 그래프, 함수


- 이산확률 변수
      countably infinite 변수
      확률 질량 함수를 통해 확률을 얻을 수 있다.

  - 확률 질량 함수

      P(a ≤ X ≤ b) = 𝚺f(x)
      


- 연속확률 변수
      하나로 특정할 수 없고 셀수 없는 변수
    - 확률 밀도 함수
          확률 밀도 함수 f(x)는 연속확률 변수 X에대한 확률에 분포를 나타내는 함수
          모든 x에 대해서 f(x) ≥ 0
          확률 밀도 함수의 전체 면적은 1
          특정 구간에서의 확률은 확률밀도 함수를 적분한 값이다.

## 정규 분포
- 정의
      확률 변수 X가 평균이 m이고 분산이 r인 정규분포는 X ~ N(m, r)

- 특징
      모든 정규분포는 좌우 대칭
      정규곡선의 위치는 평균에 의해서 결정된다.
      표본공간에 따라 평균과 표준편차는 다르지만 평균으로부터 k배 표준편차 이내에 확률 변수 X가 존재할 확율을 동일

## 분산과 표준편차

- 분산 확률분포함수의 모양을 설명할 수 있다.
- 분산은 확률분포함수에 확률이 모여있는지 퍼져있는지를 알 수 있는 수치적 파라미터

$${σ^2} = Var[X] = E[{(X-\mu)^2}]$$

- 성질
      Var[X] >= 0
      Var[cX] = c^2Var[X]

      Var[X+Y] = Var[X] + Var[Y] + 2E[(X-m_x)(Y-m_y)]



- 표본평균의 분산과 표본의 분산의 관계
$$E[X] = E[\bar{X}]$$
$$Var[\bar{X}] = \frac{1}{N}Var[X]$$
