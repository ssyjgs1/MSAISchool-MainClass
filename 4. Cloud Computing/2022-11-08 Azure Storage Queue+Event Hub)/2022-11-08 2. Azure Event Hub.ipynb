```python
# 이벤트 관련 패키지 불러오기
# 데이터를 이벤트 허브쪽으로 쏠 수 있게 함
from azure.eventhub.aio import EventHubProducerClient
from azure.eventhub import EventData
```

```python
# 연결 문자열로 연결하겠다
# Azure > eventhub > event1 > shared access policies > sas rule
eventhub = EventHubProducerClient.from_connection_string(
    conn_str='ㅁㅁㅁㅁㅁㅁㅁㅁ')
```

```python
# 일정 주기로 시간 정보 보내기
import time
from datetime import datetime

i=0
while(1): # 1은 True이므로 항상 참이 성립됨
    message = 'time : ' + str(datetime.now())
    
    # 전송하기
    # send_event : 단건 / send_batch : 복수건
    eventhub.send_event(EventData(message))

    # 20초에 한 번씩 메시지 출력
    i += 1
    if i%10==0 :
        print('message {}th sending...'.format(i))    
    
    time.sleep(2)
```
