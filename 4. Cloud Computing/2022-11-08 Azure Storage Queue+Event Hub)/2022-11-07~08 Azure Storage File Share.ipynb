```python
# File Share 연결하는 코드 부분
from azure.storage.file import FileService
file_service = FileService(account_name='ㅁㅁㅁㅁ',
                           account_key='ㅁㅁㅁㅁㅁㅁㅁㅁ')
```

```python
# myshare 공유 폴더를 생성
file_service.create_share('myshare')
```

```python
# myshare 안에 sampledir이라는 샘플 디렉토리 생성
file_service.create_directory('myshare','sampledir')
```

```python
# 파일 업로드 할 때 컨텐츠 설정을 위한 패키지
# jgm.jpg를 myfile로 복사함
from azure.storage.file import ContentSettings

file_service.create_file_from_path('myshare', # 경로명
                                   None, # 하위 폴더 지정 안 함
                                   'myfile', # 올릴 파일 이름 지정
                                   'jgm.jpg',
                                   content_settings=ContentSettings(content_type='image/jpg'))
```

```python
# 공유 폴더 내부의 파일 목록 가져와서 보기
generator = file_service.list_directories_and_files('myshare')
for file_or_dir in generator:
    print(file_or_dir.name)
```

```python
# 특정 경로에 파일 가져오기
# 다시 myfile로 복사된 jgm.jpg를 outjgm.jpg로 가져오는 과정
# 경로에서, 하위경로 미선택, 불러올 파일, 갖고온 새이름
file_service.get_file_to_path('myshare',None,'myfile','outjgm.jpg')
```

```python
# 스냅샷 찍어두는 부분
metadata = {"foo":"bar"} # 태그를 달아둠(필요한 공유 파일을 찾기 위한 용도)
snapshot = file_service.snapshot_share('myshare',metadata=metadata)
```

```python
# 전체 스냅샷 목록 만듦
shares = list(file_service.list_shares(include_snapshots=True))

for s in shares:
    print(s.snapshot)
```

```python
# 스냅샷의 파일과 디렉토리 목록 가져옴
directories_and_files = list(
    file_service.list_directories_and_files('myshare',snapshot='2022-11-08T01:07:58.0000000Z')
)
```

```python
# 파일과 디렉토리 목록 출력
directories_and_files
for file_or_dir in generator:
    print(file_or_dir.name)
```

```python
# 스냅샷 삭제
file_service.delete_share('myshare',snapshot='2022-11-08T01:07:58.0000000Z')
```

```python
# 공유 폴더 내부의 파일 삭제
file_service.delete_file('myshare',None,'myfile')
```

```python
# 공유 폴더를 통째로 삭제
file_service.delete_share('myshare')
```
