```python
import requests
from PIL import Image, ImageDraw, ImageFont
from io import BytesIO
```

```python
subscription_key='ㅁㅁㅁㅁㅁㅁㅁㅁㅁㅁㅁ'
vision_base_url='ㅁㅁㅁㅁㅁㅁㅁㅁㅁㅁㅁㅁㅁㅁ'
ocr_url = vision_base_url + 'ocr'
```

```python
image_url = "https://img1.daumcdn.net/thumb/R1280x0/?fname=http://t1.daumcdn.net/brunch/service/user/Me7/image/v7jUT-b22_3e9Te6Fcz8U3_xhwQ.jpg"
```

```python
image = Image.open(BytesIO(requests.get(image_url).content))
image
```

```python
headers = {'Ocp-Apim-Subscription-Key' : subscription_key}
params = {'language':'unk','detectOrientation':'true'}
data = {'url':image_url}
```

```python
response = requests.post(ocr_url,headers=headers,params=params,json=data)
```

```python
result = response.json()
result
```

```python
for region in result['regions']:  #json 구조에서 원하는 값을 뽑아내는 방법
  lines = region['lines']

  for line in lines:
    words = line['words']

    for word in words:
        print(word['text'])
```
